```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('employees.csv')
df
```

```python
sns.histplot(df['Salary'])
plt.show()
```

```python
sns.histplot(df['Salary'], kde=True, bins=10)
plt.title("Salary Distribution")
plt.xlabel("Salary")
plt.ylabel("Frequency")
plt.show()
```

```python
sns.boxplot(
           data=df,
            x='Team',
            y='Salary')
plt.title("Salary Distribution by Team")
plt.xlabel("Team")
plt.ylabel("Salary")
plt.figure(figsize=(10,10))
plt.show()
```

```python
sns.violinplot(
           data=df,
            x='Team',
            y='Salary',
           hue='Team',
           split=True)
plt.title("Salary Distribution by Team")
plt.xlabel("Team")
plt.ylabel("Salary")

plt.show()
```

```python
sns.barplot(df['Team'])
plt.show()
```

```python
sns.scatterplot(
    data=df,
    x='Salary',
    y='Bonus %'
)
plt.title("Salary vs Bonus %")
plt.xlabel("Salary")
plt.ylabel("Bonus %")
plt.show()
```

```python
sns.pairplot(df)
plt.show()
```

```python
sns.barplot(
    data=df,
    x='Team',
    y='Salary',
    hue='Gender'
)
plt.title("Salary by Team")
plt.xlabel("Team")
plt.ylabel("Salary")
plt.show()
```
