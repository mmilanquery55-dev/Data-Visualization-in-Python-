```python
import matplotlib.pyplot as plt
x=[1,2,3,4]
y=[5,6,7,8]
plt.plot(x,y)
plt.show()
```

```python
import seaborn as sns
sns.lineplot(x=[1,2,3,4],y=[5,6,7,8])
plt.show()
```

```python
df=sns.load_dataset('tips')
df.head()
```

```python
sns.scatterplot(
    data=df,
    x='total_bill',
    y='tip',
    hue='sex',
    size='time'
)

plt.show()
```

```python
sns.scatterplot(
    data=df,
    x='total_bill',
    y='tip',
    # hue='sex',
    # size='time',
    style='smoker'
)

plt.show()
```
