```python
import matplotlib.pyplot as plt
x=[1,2,3,4,5]
y=[10,20,24,35,50]
plt.plot(x, y)
plt.show()
```

Pie Chart

```python
# Used to show part to whole relationships/data
import matplotlib.pyplot as plt
categories = ['A','B','C','D','E']
sales = [10,20,26,36,45]
plt.pie(sales,
        labels=categories,
        autopct='%1.1f%%',
        startangle=90)
plt.title("Sales Pie Chart")
plt.show()
```

Scatter Plot

```python
# used to find relationship between variables
# data
y1=[10,15,20,25,30]
y2=[15,19,23,35,42]
plt.scatter(y1,y2)
plt.title("Example of Scatter Plot")
plt.show()
```

SubPlots

```python
# used to show multiple charts in one figure
# data -1 -> bar chart
categories = ['MOn','Tue','Wed','Thu','Fri']
sales = [10,20,26,36,45]

# data -2 -> Scatter plot

y1=[10,15,20,25,30]
y2=[15,19,23,35,42]

plt.figure(figsize=(10,4))

# first plot-bar chart
# row , columns and position
plt.subplot(1,2,1)
plt.bar(categories,sales )
plt.title("Daily Sales")
plt.xlabel("Days")
plt.ylabel("Sales")

# Second scatter chart
plt.subplot(1,2,2)
plt.scatter(y1,y2)
plt.title("User Example")
plt.xlabel("user 1")
plt.ylabel("user 2")
plt.show()
```

Matplotlib with pandas - real data

```python
# create data frame
import pandas as pd
data ={
    'Month':['Jan','Feb','Mar','Apr'],
    'Sales' :[1200,1100,1300,1500]
}
df=pd.DataFrame(data)
df
```

```python
plt.bar(df['Month'],df['Sales'])
plt.title("Matplotlib with pandas")
plt.xlabel('Month')
plt.ylabel('Sales')
plt.show()
```

Save charts / plot


```python
plt.bar(df['Month'],df['Sales'])
plt.title("Matplotlib with pandas")
plt.xlabel('Month')
plt.ylabel('Sales')

plt.savefig("Monthly_Sales_From_df.png")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

plt.style.use('_mpl-gallery-nogrid')

# make data: correlated + noise
np.random.seed(1)
x = np.random.randn(5000)
y = 1.2 * x + np.random.randn(5000) / 3

# plot:
fig, ax = plt.subplots()

ax.hexbin(x, y, gridsize=20)

ax.set(xlim=(-2, 2), ylim=(-3, 3))

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

plt.style.use('_mpl-gallery')

# make data:
np.random.seed(10)
D = np.random.normal((3, 5, 4), (1.25, 1.00, 1.25), (100, 3))

# plot
fig, ax = plt.subplots()
VP = ax.boxplot(D, positions=[2, 4, 6], widths=1.5, patch_artist=True,
                showmeans=False, showfliers=False,
                medianprops={"color": "white", "linewidth": 0.5},
                boxprops={"facecolor": "C0", "edgecolor": "white",
                          "linewidth": 0.5},
                whiskerprops={"color": "C0", "linewidth": 1.5},
                capprops={"color": "C0", "linewidth": 1.5})

ax.set(xlim=(0, 8), xticks=np.arange(1, 8),
       ylim=(0, 8), yticks=np.arange(1, 8))

plt.show()
```
